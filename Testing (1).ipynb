```python
import pandas as pd
from warpdrive import WarpDrive

wd = WarpDrive()

# Get file path (string)
data_path = wd.get_args("data")

# Convert to DataFrame
data = pd.read_csv(data_path)
```

```python
table = data.head()
wd.save_table(table, "Data Head")
```

```python
import plotly.express as px

geo_counts = data["Geography"].value_counts().reset_index()
geo_counts.columns = ["Geography", "Count"]

fig = px.bar(
    geo_counts,
    x="Geography",
    y="Count",
    title="Customers by Geography"
)

wd.save_graph(fig, "Customers by Geography")
```
