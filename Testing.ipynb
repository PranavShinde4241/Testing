```python
!pip install -r requirements.txt
from warpdrive import WarpDrive

wd = WarpDrive()
data= wd.get_args("data")

# your code here
table=data. Head()
wd.save_table(table,"Data Head")

import plotly.express as px

geo_counts = data["Geography"].value_counts().reset_index()
geo_counts.columns = ["Geography", "Count"]

fig = px.bar(
    geo_counts,
    x="Geography",
    y="Count",
    title="Customers by Geography"
)

wd.save_graph(fig, "Customers by Geography")

fig = px.scatter(
    data,
    x="Age",
    y="Balance",
    title="Age vs Balance"
)

wd.save_graph(fig, "Age vs Balance")
```
